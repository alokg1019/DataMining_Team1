```
def get_similar_anime(anime_name):
    if anime_name not in new_data.index:
        return None, None
    else:
        sim_animes = item_sim_df.sort_values(by=anime_name, ascending=False).index[1:]
        sim_score = item_sim_df.sort_values(by=anime_name, ascending=False).loc[:, anime_name].tolist()[1:]
        return sim_animes, sim_score
```

```
items=['12 ASS ZINC CHRISTMAS DECORATIONS','12 COLOURED PARTY BALLOONS']

def get_rc(arr):
  res = {}
  for i in arr:
    animes, score = get_similar_anime(i)
    for x,y in zip(animes[:10], score[:10]):
      if x in res:
        if res[x] > y:
          continue
        else:
          res[x] = y
      else:
        res[x] = y
          
  return res
    
resp = get_rc(items)
res = sorted(resp.items(), reverse=True, key=lambda x: x[1])
print(res)

```
