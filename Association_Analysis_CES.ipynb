```python
#Now splitting data according to the city of transaction 

# Transactions done in Belem 
basket_Belem = (data[data['city'] == 'Belem'].groupby(['id', 'description'])['quantity']
                .sum().unstack().reset_index()
                .fillna(0)
                .set_index('id'))
  
basket_Belem.head()
```
